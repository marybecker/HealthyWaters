# Identifying High Quality Rivers and Streams in Connecticut

### Abstract
---

The objective of the clean water act is to 'restore and maintain the chemical, physical and biological integrity of the Nation's waters.'  However, much of the work conducted to date both around the country and in Connecticut has focused on restoring impaired waters. More recent work under EPA's [healthy watersheds program](https://www.epa.gov/hwp) and [in Connecticut](https://portal.ct.gov/-/media/DEEP/water/water_quality_management/monitoringpubs/Bellucci-et-al-2011-Northeastern-Naturalist.pdf) has begun to bring more emphasis on identifying and protecting high quality waters.  These efforts are needed as [preliminary work in Connecticut](https://marybecker.github.io/Biointegrity/) noted the possible reduction of taxa highly sensitive to pollution (BCG 2 - Figure 1) at long-term sentinel stream monitoring stations over the past ~ 30 years.  

This project seeks to explore how drainage basin characteristics differ in least disturbed high quality Connecticut streams that support highly sensitive taxa compared to streams with more human alteration and .  This information will be used to identify and map the high quality streams across the State of Connecticut.  This type of information could be used to target and support protective management efforts for these streams.

### Background
---

Biological integrity is the ability of an aquatic ecosystem to support and maintain a balanced, adaptive community of organisms having a species composition, diversity, and functional organization comparable to that of natural habitats within a region. The presence, condition, numbers and types of fish, insects, algae, plants, and other organisms provide direct, accurate information about the health of water bodies. Biological assessments measure these factors and are the primary tool used to evaluate the condition of water bodies. Because biological communities are affected by all of the stressors to which they are exposed over time, bioassessments provide information on disturbances not always revealed by water chemistry or other measurements.

In this analysis, biological condition as measured by the benthic macroinvertebrate community is explored at sampling sites in Connecticut in relation to upstream drainage basin characteristics. Benthic macroinvertebrates are animals without backbones, who inhabit the bottom of rivers and streams, as well as many other waterbody types. These organisms are very well studied and have a long history of use as indicators of water quality. Certain types, including mayflies, stoneflies, and caddisflies, can survive only in the cleanest water quality conditions. Tolerant species are those that can live in highly polluted waters, moderate species can endure a moderate level of pollution, while sensitive species only occur in waters with little to no pollution. Macroinvertebrate community data are summarized into biological condition gradient (BCG) (Davies & Jackson, 2006) metrics that describe the biological condition of the stream.  A BCG 2 stream is indicative of a community in a least disturbed condition, while a BCG 5 or 6 are streams that are highly disturbed and polluted.

[Cold water streams](https://doi:10.1080/02755947.2013.855280) in Connecticut also support native and sensitive fish species.  Recent work has identified and [mapped](https://ctdeepwatermonitoring.github.io/ColdWaterHab/) these locations across the state.

### Objectives
---
- Begin to explore how high quality least disturbed stream (BCG2) basin characteristics differ from streams with more disturbance
- Identify and map where high quality (BCG2) sites and drainage basins are located in the state
- Identify where BCG2 drainages intersect with cold water drainages
- Compile data to identify areas in Connecticut that support sensitive species and should be targeted for protection efforts

### Load Data and Combine
---

```python
#Load csv files and merge
bcg = pd.read_csv('data/bcg_BugSites_HydroID.csv')  #sites with observed BCG data
hsi = pd.read_csv('data/HSI_Catchment_AllCalcs.csv') #watershed characteristics of all catchments in state
bcg_hsi = pd.merge(bcg, hsi) #merge watershed characteristics with BCG data for exploration

#Load geospatial data
bcg2_flowline = gpd.read_file('data/upstreamFlowline_BCG2.shp')#upstream flowlines from BCG2 sites
bcg2_basin = gpd.read_file('data/upstreamCatchments_BCG2.shp') #upstream drainage basin from BCG2 sites
cw_sites = gpd.read_file('https://opendata.arcgis.com/datasets/9d1c6c229b654b9587880384b138021e_0.geojson') #cold water sites
cw_basin = gpd.read_file('https://opendata.arcgis.com/datasets/9d1c6c229b654b9587880384b138021e_1.geojson') #upstream drainage basin from cold water sites
state_boundary = gpd.read_file('https://opendata.arcgis.com/datasets/0c8c5b75cf3b430bb5632712764436de_1.geojson') #state boundary
state_boundary_line = gpd.read_file('https://opendata.arcgis.com/datasets/0c8c5b75cf3b430bb5632712764436de_0.geojson') #state boundary line
ct_regbasin = gpd.read_file('https://opendata.arcgis.com/datasets/291faa00615b4c49a7354ae0644cccaf_1.geojson') #regional basin
ct_water_line = gpd.read_file('https://opendata.arcgis.com/datasets/837df941fadd4eb59b5df2c6e1269d18_0.geojson')#ct waterbodies

#subset waterbody to just get streams
print(ct_water_line.columns)
ct_streams = ct_water_line[ct_water_line['STREAM_NO']>0]
print(len(ct_water_line))
print(len(ct_streams))
```

```python
#Add column to denote if bcg 2 site is also a cw site
bcg2_sites_prj = bcg_sites_prj[bcg_sites_prj.BCG == 2]
bcg2_sites_prj = bcg2_sites_prj[['STA_SEQ','Station_Na','HydroID','BCG','geometry']]
cw_sites_prj['CWSite']=1 #Add a column to denote if a cwsite after join

bcg2_cw_site = bcg2_sites_prj.merge(cw_sites_prj[['STA_SEQ','CWSite']],on='STA_SEQ',how='left')
bcg2_cw_site['CWSite']=bcg2_cw_site['CWSite'].fillna(0)#remove NaNs and make column binary
bcg2_cw_site
```

```python
#Identify bcg2 upstream catchments that intersect with cw drainage basins
cw_basin_prj['CWBasin']=1 #Add column before join to denot CW basin
bcg2_cw_basin = gpd.sjoin(bcg2_basin_ct[['HydroID','geometry']],cw_basin_prj[['CWBasin','geometry']],how='left',op='intersects')
bcg2_cw_basin['CWBasin']=bcg2_cw_basin['CWBasin'].fillna(0)#remove NaNs and make column binary

#Checks
#print(len(bcg2_basin_ct)) 
#print(len(bcg2_cw_basin))
#bcg2_cw_basin.CWBasin.describe()


bcg2_cw_basin = bcg2_cw_basin.dissolve(by='CWBasin') 
bcg2_cw_basin['CW']=bcg2_cw_basin.index #Add CWBasin back in as a column
bcg2_cw_basin.plot(column='CW')#Check
```

### Export Data for Web Map
---

```python
#Smooth out small gaps in the polygon
#https://gis.stackexchange.com/questions/120286/removing-small-polygon-gaps-in-shapely-polygon/120314#120314
eps = 1
bcg2_cw_basin_smooth = gpd.GeoDataFrame(bcg2_cw_basin.buffer(eps, 1, join_style=JOIN_STYLE.mitre).buffer(-eps, 1, join_style=JOIN_STYLE.mitre))
bcg2_cw_basin_smooth['CW']=bcg2_cw_basin.index
bcg2_cw_basin_smooth = bcg2_cw_basin_smooth.rename(columns={0:'geometry'})
bcg2_cw_basin_smooth = bcg2_cw_basin_smooth.set_crs(epsg=2234)
bcg2_cw_basin_smooth.head()

#Simplify the polygons to make smaller file size for web map
bcg2_cw_basin_simplify = gpd.GeoDataFrame(bcg2_cw_basin_smooth.simplify(50))
bcg2_cw_basin_df = bcg2_cw_basin_smooth[['CW']].merge(bcg2_cw_basin_simplify,left_index=True,right_index=True)
bcg2_cw_basin_df = bcg2_cw_basin_df.rename(columns={0:'geometry'})
crs = {'init' :'epsg:2234'}
bcg2_cw_basin_df = gpd.GeoDataFrame(bcg2_cw_basin_df,crs=crs,geometry='geometry')

# state_boundary_mainland = state_boundary_line[state_boundary_line['COASTA_COD']<5]
# state_boundary_simplify = gpd.GeoDataFrame(state_boundary_mainland.simplify(10))
# state_boundary_simplify = state_boundary_simplify.rename(columns={0: 'geometry'})
# state_boundary_simplify = gpd.GeoDataFrame(state_boundary_simplify,crs = {'init' :'epsg:4326'},geometry='geometry')

#Change the crs to WGS84
bcg2_cw_site_prj = bcg2_cw_site.to_crs(epsg=4326)
bcg2_cw_basin_df_prj = bcg2_cw_basin_df.to_crs(epsg=4326)

#Export out files for web map
bcg2_cw_basin_df_prj.to_file('exportData/bcg2_cw_basin.geojson',driver='GeoJSON')
bcg2_cw_site_prj.to_file('exportData/bcg2_cw_sites.geojson',driver='GeoJSON')
state_boundary_line[state_boundary_line['COASTA_COD']<5].to_file('exportData/StateBoundary.geojson',driver='GeoJSON')
```
